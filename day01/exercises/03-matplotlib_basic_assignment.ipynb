## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
#requiement and data
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
company_sales = pd.read_csv('../data/company_sales_data.csv')
company_sales
```

```python
# Write your code here
fig = plt.gcf()
fig.set_size_inches(10, 4.5)

plt.plot(company_sales['month_number'],company_sales['total_profit'])
plt.xlabel("Month number")
plt.ylabel("profit in dollar")
plt.title("company profit per month")
plt.show()
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

<img src="../data/company_sales_data_of_last_year.png" alt="company_sales_data_of_last_year" style="width: 700px;"/>


```python
# write your code here
fig = plt.gcf()
fig.set_size_inches(12.5, 4.5)

plt.plot(company_sales['month_number'],company_sales['total_profit'],'r--',
         linewidth=3,marker = 'o',mec = 'black', label='profit of last year')


plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.title("company profit per month")

plt.legend(loc="lower right")


# Show the figure
plt.show()
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
# write your code here
fig = plt.gcf()
fig.set_size_inches(10, 5.5)

plt.plot(company_sales['month_number'],company_sales['facecream'],
         marker = 'o',mec = 'black', label='face cream sales data')

plt.plot(company_sales['month_number'],company_sales['facewash'],
        marker = 'o',mec = 'black', label='face wash sales data')

plt.plot(company_sales['month_number'],company_sales['toothpaste'],
         marker = 'o',mec = 'black', label='toothpaste sales data')

plt.plot(company_sales['month_number'],company_sales['bathingsoap'],
         marker = 'o',mec = 'black', label='bathingsoap sales data')

plt.plot(company_sales['month_number'],company_sales['shampoo'],
         marker = 'o',mec = 'black', label='shampoo sales data')

plt.plot(company_sales['month_number'],company_sales['moisturizer'],
         marker = 'o',mec = 'black', label='moisturizer sales data')

plt.xlabel("Month number")
plt.ylabel("Sold units in number")
plt.title("company profit per month")

plt.legend(loc="upper left")

plt.show()
```

## Exercise 4

Also, add a grid in the plot. gridline style should .

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# write your code here
fig = plt.gcf()
fig.set_size_inches(10, 4.5)


barWidth = 0.25
 
# set heights of bars
bars2 = company_sales['facewash']
bars1 = company_sales['facecream']
 
# Set position of bar on X axis
r1 = company_sales['month_number']
r2 = [x + barWidth for x in r1]
 
# Make the plot
plt.bar(r1, bars1, width=barWidth, edgecolor='white', label='face cream sales data')
plt.bar(r2, bars2, width=barWidth, edgecolor='white', label='face wash sales data')
plt.grid(True,linewidth=0.5,color='#aaaaaa',linestyle='--')

plt.xlabel("Month number")
plt.ylabel("Sold units in number")
plt.title("facecream and facewash sales data")

plt.show()
```
